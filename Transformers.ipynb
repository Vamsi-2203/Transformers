```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
tokenizer = GPT2Tokenizer.from_pretrained('gpt2-large')
model = GPT2LMHeadModel.from_pretrained('gpt2-large', pad_token_id=tokenizer.eos_token_id)
```

```python
tokenizer.eos_token_id
```

```python
sent = " I love my Country"
input_ids = tokenizer.encode(sent, return_tensors='pt')
```

```python
input_ids
```

```python
tokenizer.decode(input_ids[0][1])
```

```python
tokenizer.decode(input_ids[0][3])
```

```python
result = model.generate(input_ids, max_length=100, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
```

```python
result
```

```python
tokenizer.decode(result[0], skip_special_tokens=True)
```
